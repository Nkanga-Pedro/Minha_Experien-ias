# Pre-processamento de dados

```python
# Escolher as colunas e renomear
livros=livros[['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher']]
livros.rename(columns={'Book-Title':'titulo','Book-Author':'autor','Year-Of-Publication':'ano','Publisher':'editora'},inplace=True)
```

```python
livros.head()
```

```python
# Escolher as colunas e renomear
usuario.rename(columns={'User-ID':'codus','Location':'morada','Age':'idade'},inplace=True)
rating.rename(columns={'User-ID':'codus','Book-Rating':'avaliacao'},inplace=True)
```

```python
rating.head()
```

```python
# Quantidade de avaliacao por usuario
rating['codus'].value_counts()
```

```python
# Livro que tenham mais de 200 avaliacoes
x=rating['codus'].value_counts()>100
x
```

```python
# Quantidade de usuario que avaliaram
y=x[x].index
print(y.shape)
```

```python
# Decisão do negocio - trazendo so usuario que avaliaram
rating=rating[rating['codus'].isin(y)]
```

```python
# Juntar as tabelas
rating_livro=rating.merge(livros,on='ISBN')
rating_livro.head()
```

```python
# Qte de avaliacao por livro
num_rating=rating_livro.groupby('titulo')['avaliacao'].count().reset_index()
```

```python
num_rating
```

```python
# Renomear o campo rating
num_rating.rename(columns={'avaliacao':'num_aval'},inplace=True)
```

```python
num_rating
```

```python
# Juntar a tabela num_rating e livro_avaliacao
final_rating=rating_livro.merge(num_rating,on='titulo')
final_rating
```

```python
# Decisao do negocio - livro que foram avaliado mais de 50 vezes
final_rating=final_rating[final_rating['num_aval']>=50]
final_rating.shape
```

```python
final_rating.drop_duplicates(['codus','titulo'],inplace=True)
final_rating.shape
```

```python
final_rating.head()
```

```python
# pivot table : transpor a linha em coluna
livro_pivot=final_rating.pivot_table(columns='codus',index='titulo',values='avaliacao')
```

```python
livro_pivot
```

```python
livro_pivot.shape
```

```python
livro_pivot.fillna(0,inplace=True)
```

```python
livro_pivot
```

```python
# Utilizando o algoritmo KNN - distancia euclidiana
# Transformar em amtriz esparsa
from scipy.sparse import csr_matrix
```

```python
livro_sparse=csr_matrix(livro_pivot)
```

```python
livro_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
model.fit(livro_sparse)
```

```python
livro_pivot
```

```python
# Testando o exemplo
distances,sugestions=model.kneighbors(livro_pivot.iloc[0,:].values.reshape(1,-1))
```

```python
# Fazendo a predicao e sugestao do livro - filme com titulo 1984
for i in range(len(sugestions)):
    print(livro_pivot.index[sugestions[i]])
```
