## Resvolvendo as questões

   *1- Com recurso à API “Text Search” devolva os 5 primeiros resultados de uma query à sua escolha*

- Este código em Python utiliza a biblioteca **requests** para fazer uma consulta à API "Text Search" do arquivo.pt e retornar os 5 primeiros resultados da consulta.
- O código começa importando a biblioteca requests e a exceção Timeout da mesma biblioteca. Em seguida, é definida a URL da API do arquivo.pt em url_api e a consulta desejada em query neste caso **Angola**. Também é definido um cabeçalho para a **requisição HTTP em headers**.
- O loop for é usado para realizar a consulta 5 vezes. Em cada iteração do loop, a biblioteca requests faz uma solicitação **HTTP GET** à API do arquivo.pt com a consulta definida anteriormente em **payload**. O cabeçalho definido em headers também é incluído na solicitação. Além disso, um tempo limite de 50 segundos é definido para a solicitação.

- Se a consulta for bem-sucedida, a URL da consulta é impressa na saída padrão do console. Se ocorrer um erro de tempo limite, a mensagem **"Timeout has been raised."** é impressa no console.

```python
import requests
from requests.exceptions import Timeout

# Define a URL da API do arquivo.pt e a consulta Angola
url_api = "https://arquivo.pt/textsearch"
query = "Angola"
# Define um cabeçalho para a solicitação HTTP
headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:100.0) Gecko/20100101 Firefox/100.0 "}
# Define os parâmetros da consulta como um dicionário
payload = {'q': query}
# Loop que faz a consulta 5 vezes
for i in range(5):
    try:
      # Faz a solicitação HTTP GET à API do arquivo.pt
        r = requests.get(url_api, params=payload, headers=headers, timeout = 50)
        # Imprime a URL da consulta
        print(f"Query {i+1}: {r.url}")
    except Timeout:
      # Trata o erro de tempo limite
        print('Timeout has been raised.')
```

   *2- Imprima o JSON devolvido em formato Pretty Print.*

- Este código é responsável por obter o conteúdo JSON retornado pela API após a chamada HTTP.

```python
content = r.json()
```

- O código abaixo é responsável por imprimir de forma mais legível os resultados obtidos a partir de uma consulta à **API do arquivo.pt** utilizando a biblioteca **pprint**. O conteúdo retornado pela consulta é armazenado na variável "content" como um **objeto JSON**, que é então passado como argumento para o método **pprint.pprint**, que imprime o conteúdo formatado de forma mais legível para o usuário final. Neste caso específico, apenas a lista de **response_items** é impressa

```python
#Impressao com recurso ao pretty print - mais entendivel
import pprint
pprint.pprint(content['response_items'])
```

  *3- Percorra cada um dos 5 resultados anteriormente obtidos e imprima o título da página web (title), o url (linkToArchive), o timestamp(tstamp) e o respetivo conteúdo (linkToExtractedText).*

- O Código percorre a lista de resultados de uma pesquisa na **API do site arquivo.pt**, imprimindo as informações mais relevantes dos 5 primeiros resultados.
- A variável **counter** é inicializada como 0, e é usada para controlar o número de resultados impressos. O loop "for" percorre cada elemento na lista **response_items** dentro do dicionário **content**.
- Dentro do loop, cada informação relevante de cada resultado é impressa na tela, incluindo o **título**, **link**, **data** e **conteúdo extraído do link**. Depois de imprimir as informações de um resultado, o contador é incrementado e o próximo resultado é impresso até que o contador chegue a 5.

```python
# Inicializa a variável counter como 0
counter = 0
# Itera sobre cada elemento na lista 'response_items'
for element in content['response_items']:
    if counter == 5:  # Verifica se a variável counter já é igual a 5
        break
    # Imprime o título, link, data e conteúdo do elemento
    print("Titulo: ",element['title'])
    print("Link: ",element['linkToArchive'])
    print("Time: ",element['tstamp'])
    print("Contents: ",element['linkToExtractedText'])
    print("\n")
    # Incrementa a variável counter em 1 a cada iteração do loop
    counter += 1
```

O código é utilizado para salvar os resultados da pesquisa em cache no **Google Drive**. Ele utiliza a biblioteca **os** para criar o diretório para armazenar o arquivo **JSON**, a biblioteca **datetime** para gerar um nome de arquivo único e a biblioteca json para escrever e ler dados JSON.

As etapas para executar essa tarefa:
- Monta o Google Drive
- Define o nome do arquivo de cache
- Cria o diretório de cache, se necessário
- Salva os resultados em um arquivo JSON no Google Drive
- Carrega o arquivo JSON do Google Drive
- Imprime a mensagem de sucesso

```python
import os
import datetime
import json
from google.colab import drive

# Monta o Google Drive
drive.mount('/content/drive')

# Define o nome do arquivo de cache
now = datetime.datetime.now()
filename = f"/content/drive/MyDrive/data/{now.year}.{now.month:02d}.{now.day:02d}/{query}.json"

# Cria o diretório de cache, se necessário
os.makedirs(os.path.dirname(filename), exist_ok=True)

# Salva os resultados em um arquivo JSON no Google Drive
with open(filename, "w") as f:
    json.dump(content, f)

# Carrega o arquivo JSON do Google Drive
with open(filename, "r") as f:
    cached_results = json.load(f)

# Imprime a mensagem de sucesso
print(f"Resultados salvos em cache no arquivo {filename}")
```

Por fim este código cria um diretório com a data atual (no formato "aaaa.mm.dd") dentro do Google Drive, caso ele ainda não exista. Em seguida, ele cria um arquivo chamado "query.json" dentro desse diretório e grava nele os resultados em cache obtidos através da API do arquivo.pt, utilizando o módulo json.

```python
# Cria o diretório /data/aaaa.mm.dd
today = datetime.datetime.now().strftime('%Y.%m.%d')
directory = '/content/drive/My Drive/data/{}'.format(today)
if not os.path.exists(directory):
        os.makedirs(directory)
 # Grava os resultados em cache em um arquivo JSON
filename = 'query.json'
filepath = os.path.join(directory, filename)
with open(filepath, 'w') as file:
    json.dump(content, file)
```
